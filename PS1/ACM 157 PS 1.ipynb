```python
import matplotlib.pyplot as plt
import numpy as np
import pylab 
import scipy.stats as stats
```

```python
# Problem 4a

xs = np.random.normal(size=15)

# Plot normal-quantile plot
stats.probplot(xs, dist="norm", plot=pylab)
pylab.title('Normal-quantile plot (n = 15)')
pylab.show()

# Plot histogram
plt.hist(xs)
plt.title('Histogram of sample from N(0, 1) (n = 15)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
# The points on the QQ plot are not exactly in a straight line.
# The histogram above is approximately symmetric and bell-shaped, but not necessarily unimodal
# (although a major peak is in the middle), but this changes every time a new sample is drawn.
# The points on the QQ plot deviate more from linearity the less symmetric, unimodal,
# and bell-shaped the histogram is.
```

```python
# Problem 4b

# Plots normal-quantile plot and histogram from a sample of size n drawn from N(0, 1)
def plotNorm(n):
    xs = np.random.normal(size=n)

    # Plot normal-quantile plot
    stats.probplot(xs, dist="norm", plot=pylab)
    pylab.title('Normal-quantile plot (n = {})'.format(n))
    pylab.show()

    # Plot histogram
    plt.hist(xs)
    plt.title('Histogram of sample from N(0, 1) (n = {})'.format(n))
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.show()
    
plotNorm(50)
plotNorm(100)
plotNorm(1000)
```

```python
# The points on the QQ plot approximately fall in a straight line when n = 1000,
# and less so when n is smaller. The histogram is also roughly symmetric, unimodal,
# and bell-shaped when n = 1000 and less so when n is smaller.
```

```python
# Problem 4c

# My estimate for the "critical" sample size n* is about 1000, since the points
# on the QQ plot seem to consistently fall in a roughly straight line when n is at least 1000,
# which is due to the fact that the histogram is consistently symmetric, unimodal,
# and bell-shaped when n is at least 1000.
```

```python
# Problem 5a

data = np.loadtxt('birth.txt')

motherHeights = data[:,4]
motherHeights = np.extract(motherHeights != 99, motherHeights)
plt.hist(motherHeights, bins=20)
plt.title('Histogram of mothers\' heights (inches)')
plt.xlabel('Heights (inches)')
plt.ylabel('Frequency')
plt.show()

# Having 20 bins seems optimal for representing the shape of the distribution 
# (unimodal, bell-shaped, roughly symmetric) smoothly without gaps.
```

```python
# Problem 5b

mean = np.mean(motherHeights)
print('mean = ', mean)
print('median = ', np.median(motherHeights))
std = np.std(motherHeights, ddof=1)
print('standard deviation = ', std)
print('IQR = ', stats.iqr(motherHeights))

# The center of the sample is well-defined, as the mean and median are very close in value.
```

```python
# Problem 5c

plt.boxplot(motherHeights)
plt.ylabel('Height (inches)')
plt.title('Boxplot of mothers\' heights (inches)')
plt.show()

plt.plot(np.sort(motherHeights), np.linspace(0, 1, len(motherHeights), endpoint=False))
plt.title('eCDF of mothers\' heights (inches)')
plt.xlabel('Height (inches)')
plt.ylabel('Cumulative probability')
plt.show()

stats.probplot(motherHeights, dist='norm', plot=pylab)
pylab.title('QQ plot for mothers\' heights')
pylab.show()

# The sample seems to be approximately normal with mean of 64.05 and variance of 6.42,
# as the histogram is unimodal, bell-shaped, roughly symmetric, the boxplot also
# shows symmetry excluding outliers, the eCDF resembles CDF of a normal disribution,
# and the points on the QQ plot fall in line (except the first few which are outliers).
```

```python
# Problem 5d

smokeHeights = np.extract(data[:,6] == 1, data[:,4])
smokeHeights = np.extract(smokeHeights != 99, smokeHeights)
noSmokeHeights = np.extract(data[:,6] == 0, data[:,4])
noSmokeHeights = np.extract(noSmokeHeights != 99, noSmokeHeights)

plt.boxplot([smokeHeights, noSmokeHeights])
plt.title('Boxplots of heights of mothers who smoked / did not smoke (inches)')
plt.xticks([1, 2], ['Smoke', 'No smoke'])
plt.ylabel('Height')
plt.show()

# It's hard to conclude that one group is higher on average than the other, 
# since the medians of both groups are around 64 inches.
```
